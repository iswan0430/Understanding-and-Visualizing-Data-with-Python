```python
import seaborn as sns
import matplotlib.pyplot as plt

tips_data = sns.load_dataset("tips")
```

<a name="1"></a>
### Creating a Histogram

```python
tips_data.head()
```

```python
tips_data.head(2)
```

```python
tips_data.describe()
```

```python
sns.distplot(tips_data["total_bill"], kde=False).set_title("Histogram of total bills")
plt.show()
```

```python
sns.distplot(tips_data["tip"], kde=False).set_title("Histogram of Tips")
plt.show()
```

```python
sns.distplot(tips_data["total_bill"], kde=False)
sns.distplot(tips_data["tip"], kde=False).set_title("Histogram of Total Bill and Tip")
plt.show()
```

<a name="2"></a>
### Creating a Boxplot

```python
sns.boxplot(tips_data["total_bill"])
sns.boxplot(tips_data["tip"]).set_title("Box Plot of Total Bill and Tip")
plt.show()
```

<a name="3"></a>
### Creating Histograms and Boxplots Plotted by Groups

```python
sns.boxplot(x=tips_data["tip"],y=tips_data["time"])

g = sns.FacetGrid(tips_data, row = "time")
g = g.map(plt.hist, "tip")

plt.show()
```

```python
sns.boxplot(x=tips_data["tip"],y=tips_data["sex"])

s = sns.FacetGrid(tips_data, row = "sex")
s = s.map(plt.hist, "tip")

plt.show()
```

```python
ss = sns.FacetGrid(tips_data, row="size")
ss = ss.map(plt.hist, "tip")

plt.show()
```
